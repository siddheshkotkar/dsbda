```python
words_without_stop_words_data1 = []
for word in data1:
    if word.text not in stopwords.words():
        words_without_stop_words.append(word.text)
```

```python
words_without_stop_words_data1 = [word.text for word in data1 if word.text not in stopwords.words()]
```

```python
len(words_without_stop_words_data1)
```

```python
words_without_stop_words_data2 = [word.text for word in data2 if word.text not in stopwords.words()]
len(words_without_stop_words_data2)
```

```python
ps = PorterStemmer()
stemmed_data1 = []
for word in words_without_stop_words_data1:
    stemmed_data1.append(ps.stem(word))
```

```python
ps = PorterStemmer()
stemmed_data2 = []
for word in words_without_stop_words_data2:
    stemmed_data2.append(ps.stem(word))
```

```python
wl = WordNetLemmatizer()
lemmatized_data1 = []
for word in words_without_stop_words_data1:
    lemmatized_data1.append(wl.lemmatize(word))
```

```python
wl = WordNetLemmatizer()
lemmatized_data2 = []
for word in words_without_stop_words_data2:
    lemmatized_data2.append(wl.lemmatize(word))
```

```python
tf = []
corpus = [lemmatized_data1,lemmatized_data2]
for data in corpus:
    count = {}
    for word in data:
        if word in count:
            count[word] += 1
        else:
            count[word] = 1
    tf.append(count)
```

```python
for i in range(len(corpus)):
    length = len(corpus[i])
    for words in tf[i].keys():
        tf[i][words] = tf[i][words] / length
```

```python
corpus = [lemmatized_data1,lemmatized_data2]
unique_words = set(lemmatized_data1).union(set(lemmatized_data2))
idf = {}
for word in unique_words:
    idf[word] = 0
    for data in corpus:
        if word in data:
            idf[word]+=1
```

```python
length_of_corpus = len(corpus)
for word in idf.keys():
    idf[word] = log10(length_of_corpus / idf[word])
```

```python
tf_idf = []
for sentance in tf:
    dict_sent = {}
    for word in sentance.keys():
        dict_sent[word] = sentance[word] * idf[word]
    tf_idf.append(dict_sent)
```
