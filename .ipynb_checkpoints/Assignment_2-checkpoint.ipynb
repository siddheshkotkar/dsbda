```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance_modified.csv")
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df['math score'] = pd.to_numeric(df['math score'], errors='coerce')
```

```python
df['math score'].fillna(df['math score'].mean(), inplace=True)
```

```python
df['reading score'].fillna(df['reading score'].mean(), inplace=True)
df['writing score'].fillna(df['writing score'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
for cols in df.columns:
    if df[cols].dtype=='object':
        df[cols]=LabelEncoder().fit_transform(df[cols])
    
```

```python
df.head()
```

```python
scaler = MinMaxScaler()
```

```python
df[['math score','reading score','writing score']] = scaler.fit_transform(df[['math score','reading score','writing score']])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
q1 = df.quantile(.25)
q3 = df.quantile(.75)
```

```python
IQR = q3 - q1
```

```python
ll = q1 - (1.5*IQR)
```

```python
ul = q3 + (1.5*IQR)
```

```python
outliars1 = df < ll
```

```python
outliars1.sum()
```

```python
df = df[~np.any(outliars1, axis=1)]
```

```python
df[['math score', 'reading score', 'writing score']] = StandardScaler().fit_transform(df[['math score', 'reading score', 'writing score']])
```
