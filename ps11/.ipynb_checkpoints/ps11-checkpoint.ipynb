```python
df=sns.load_dataset('titanic')
```

```python
df.drop('deck',axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
# df['embarked'].fillna(df['embarked'].mean(),inplace=True)
# df['embark_town'].fillna(df['embark_town'].mean(),inplace=True)
```

```python
df['embarked']
```

```python
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
```

```python
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
plt.title('Alive Boxplot')
sns.boxplot(data=df[df['alive'] == 'yes'], x='sex' , y='age')
```

```python
plt.title('Dead Boxplot')
sns.boxplot(data=df[df['alive'] == 'no'], x='sex',y='age')
```
