```python
import docx2txt
import spacy
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from math import *
```

```python
text=docx2txt.process('testdoc.docx')
text2=docx2txt.process('testdoc2.docx')
```

```python
nlp=spacy.load('en_core_web_md')
```

```python
data=nlp(text)
```

```python
data2=nlp(text2)
```

```python
for word in data:
    print(f"{word.text} --> {word.pos_}")
```

```python
for word in data2:
    print(f"{word.text} --> {word.pos_}")
```

```python
words_without_stop_words_data1=[]
for word in data:
    if word.text not in stopwords.words():
        words_without_stop_words_data1.append(word.text)
```

```python
words_without_stop_words_data2=[]
for word in data2:
    if word.text not in stopwords.words():
        words_without_stop_words_data2.append(word.text)
```

```python
ps=PorterStemmer()
stemmed_data1=[]
for word in words_without_stop_words_data1:
    stemmed_data1.append(ps.stem(word))
```

```python
stemmed_data2=[]
for word in words_without_stop_words_data2:
    stemmed_data2.append(ps.stem(word))
```

```python
lm=WordNetLemmatizer()
lemmatized_data1=[]
for word in stemmed_data1:
    lemmatized_data1.append(lm.lemmatize(word))
```

```python
lemmatized_data2=[]
for word in stemmed_data2:
    lemmatized_data2.append(lm.lemmatize(word))
```

```python
tf=[]
corpus=[lemmatized_data1,lemmatized_data2]
for data in corpus:
    count={}
    for word in data:
        if word in count:
            count[word]+=1
        else:
            count[word]=1
    tf.append(count)
```

```python
for i in range(len(corpus)):
    for word in tf[i].keys():
        tf[i][word]=tf[i][word]/len(corpus[i])
```

```python
unique_words=set(lemmatized_data1).union(set(lemmatized_data2))
idf={}
for word in unique_words:
    idf[word]=0
    for data in corpus:
        if word in data:
            idf[word]+=1
```

```python
for word in idf.keys():
    idf[word]=log10(len(corpus)/idf[word])
```

```python
tf_idf=[]
for sentence in tf:
    dict_sent={}
    for word in sentence.keys():
        dict_sent[word]=sentence[word]*idf[word]
    tf_idf.append(dict_sent)
```
